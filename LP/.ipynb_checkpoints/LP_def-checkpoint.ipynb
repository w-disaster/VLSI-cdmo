```python
import pulp as pl
from pulp import *
import math
import csv
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import matplotlib.patches as mpatches
```

```python
def read_instance(num):
    
    with open('instances\ins-{}.txt'.format(num), 'r') as fd:
        rows = []
        for row in csv.reader(fd):
            for i in row:
                rows.append(i.replace(" ", ","))

    values = []

    for element in rows:
        values.append([int(i) for i in element.split(',')])
    
    data = {}
    keys = ["width", "n_circuits", "circuit_w", "circuit_h", "h_max", "max_y"]
    structured = [values[0][0], values[1][0], [], []]
    values.pop(0)
    values.pop(0)
    
    for i in values:
        x_size = int(i[0])
        y_size = int(i[1])
        structured[2].append(x_size)
        structured[3].append(y_size)

    max_y = max(structured[3])

    for key, val in zip(keys, structured):
        data[key] = val

    data['h_max'] = math.ceil(data['n_circuits']*max(data['circuit_h']) / math.floor(data['width']/max(data['circuit_w'])))
    data['max_y'] = max_y

    return data

def save_data(out, num):
    open(os.path.join("out", "ins-{}-solved.txt".format(num)), "w").write(out)
```

```python
for instance in range(1, 41):
    
    data = read_instance(instance)

    width = data['width']
    n_circuits = data['n_circuits']
    circuit_w = data['circuit_w']
    circuit_h = data['circuit_h']
    h_max = data['h_max']
    max_y = data['max_y']

    prob = LpProblem("VLSI", LpMinimize)

    solver_list = pl.listSolvers()
    path_to_cplex = r'C:\Program Files\IBM\ILOG\CPLEX_Studio_Community221\cplex\bin\x64_win64\cplex.exe'
    solver = pl.CPLEX_CMD(path=path_to_cplex, timeLimit=180)

    x = [LpVariable("x{}".format(i + 1), 0, width - min(circuit_w), LpInteger) for i in range(n_circuits)]
    y = [LpVariable("y{}".format(i + 1), 0, h_max - min(circuit_h), LpInteger) for i in range(n_circuits)]
    
    delta = [[[LpVariable("delta{}-{}-{}".format(i + 1, j + 1, k + 1), 0, 1, cat="Integer") for j in range(n_circuits)]
              for i in range(n_circuits)] for k in range(n_circuits)]
    
    height = LpVariable("Height", max_y, h_max, LpInteger)
    prob += height, "Height of the plate"

    for i in range(n_circuits):
        prob += x[i] + circuit_w[i] <= width
        prob += y[i] + circuit_h[i] <= height
        for j in range(n_circuits):
            if i < j:
                prob += x[i] + circuit_w[i] <= x[j] + width * (delta[i][j][0])
                prob += x[j] + circuit_w[j] <= x[i] + width * (delta[i][j][1])
                prob += y[i] + circuit_h[i] <= y[j] + h_max * (delta[i][j][2])
                prob += y[j] + circuit_h[j] <= y[i] + h_max * (delta[i][j][3])
                prob += delta[i][j][0] + delta[i][j][1] + delta[i][j][2] + delta[i][j][3] <= 3
                
    prob.solve(solver)

    out = ""
    out += "{} {}\n{}\n".format(width, int(height.varValue), n_circuits)

    for i in range(n_circuits):
        out += "{} {} {} {}\n".format(circuit_w[i], circuit_h[i], int(x[i].varValue), int(y[i].varValue))

    out += "solved"

    save_data(out, instance)
```

```python
def generate_cell(rows):
        w = int(rows[0][0])
        h = int(rows[0][1])
        l = int(rows[1][0])
 
        cell = np.ones((w, h))

        for i in range(l):
            for y_element in range(rows[i + 2][1]):
                for x_element in range(rows[i + 2][0]):
                    cell[rows[i + 2][2] + x_element, rows[i + 2][3] + y_element] = int(i + 1)
        cell = np.rot90(cell)
        cell = np.flipud(cell)
        return cell


for num in range(1, 41):
    with open('out\ins-{}-solved.txt'.format(num), 'r') as fd:
        rows = []
        for row in csv.reader(fd):
            if row == ['solved']:
                break
            for i in row:
                rows.append(i.replace(" ", ","))

    values = []

    for element in rows:
        values.append([int(i) for i in element.split(',')])
    
    legend = []
    for element in range(2, len(values)):
        legend.append("size: " + str(values[element][0]) + ", " + str(values[element][1]) +
                     "\ncoordinate: " + str(values[element][2]) + ", " + str(values[element][3]))

    try:
        num_colors = int(values[1][0])
        data = generate_cell(values)
        print(num)

        cmap = plt.cm.get_cmap('tab20c', num_colors)

        if num_colors >= 19:
            plt.figure(figsize = (14, 11), dpi = 100)
        else:
            plt.figure(figsize = (12, 8), dpi = 100)

        plt.pcolormesh(data, edgecolors = 'white', linewidth = 0.7, cmap = cmap)
        plt.title('Instance number {}, {}x{}'.format(num, values[0][0], values[0][1]))
        ax = plt.gca()
        ax.xaxis.set_major_locator(MaxNLocator(integer=True))
        ax.yaxis.set_major_locator(MaxNLocator(integer=True))
        ax.set_aspect('equal')
        ax.set_xlabel('Width')
        ax.set_ylabel('Height')

        bound = np.linspace(0, 1, num_colors + 1)
        bound_prep = np.round(bound * (num_colors - 1), 2)

        plt.legend([mpatches.Patch(color=cmap(b)) for b in bound[:-1]],
                   ['{}'.format(legend[i], bound_prep[i + 1] - 0.01) for i in range(num_colors)], bbox_to_anchor=(1.04, 1),
                   loc="upper left")

        plt.rcParams.update({'figure.max_open_warning': 0})
    except IndexError:
        print("Empty output")
    else:
        plt.savefig('plotted\ins-{}-plotted'.format(num))
```
